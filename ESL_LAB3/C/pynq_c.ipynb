```python
from pynq import Overlay
from pynq.lib import AxiGPIO
import time
import numpy as np
%matplotlib inline
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# 載入 bitstream
overlay = Overlay('rgb2yuv_c.bit')
c_ip = overlay.rgb2yuv_axi_0
```

```python
#讀入圖片
img = Image.open("mountain256.bmp").resize((256, 256)).convert("RGB")
r, g, b = img.split()
r = np.array(r).flatten()
g = np.array(g).flatten()
b = np.array(b).flatten()
plt.imshow(img, cmap='gray')
plt.title("RGB")
plt.show()
```

```python
# 建立輸出 buffer
c_y = np.zeros(256*256)
c_u = np.zeros(256*256)
c_v = np.zeros(256*256)
```

```python
c_start = time.time()
# 一個像素一個像素送入
for i in range(256*256):
    c_ip.write(0x10, int(r[i])) # R
    c_ip.write(0x18, int(g[i])) # G
    c_ip.write(0x20, int(b[i])) # B
    
    #time.sleep(0x0, 0.0001)
    c_ip.write(0x0,  0x01)
    
    c_y[i] = c_ip.read(0x28) # Y
    c_u[i] = c_ip.read(0x30) # U
    c_v[i] = c_ip.read(0x38) # V
c_end = time.time()
```

```python
# reshape
c_y_reshape = c_y.reshape(256, 256)
c_u_reshape = c_u.reshape(256, 256)
c_v_reshape = c_v.reshape(256, 256)
```
